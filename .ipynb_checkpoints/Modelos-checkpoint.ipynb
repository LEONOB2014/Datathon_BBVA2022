```python
import optuna
```

```python
def run(trial):
    fold = 0
    learning_rate = trial.suggest_float("learning_rate", 1e-2, 0.25, log=True)
    reg_lambda = trial.suggest_loguniform("reg_lambda", 1e-8, 100.0)
    reg_alpha = trial.suggest_loguniform("reg_alpha", 1e-8, 100.0)
    subsample = trial.suggest_float("subsample", 0.1, 1.0)
    colsample_bytree = trial.suggest_float("colsample_bytree", 0.1, 1.0)
    max_depth = trial.suggest_int("max_depth", 1, 7)

    xtrain = df[df.kfold != fold].reset_index(drop=True)
    xvalid = df[df.kfold == fold].reset_index(drop=True)

    ytrain = xtrain['Valor comercial (USD)']
    yvalid = xvalid['Valor comercial (USD)']

    xtrain = xtrain[useful_features]
    xvalid = xvalid[useful_features]

    ordinal_encoder = preprocessing.OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value=999)
    xtrain[object_cols] = ordinal_encoder.fit_transform(xtrain[object_cols])
    xvalid[object_cols] = ordinal_encoder.transform(xvalid[object_cols])

    model = XGBRegressor(
        random_state=42,
        tree_method="gpu_hist",
        gpu_id=0,
        predictor="gpu_predictor",
        n_estimators=7000,
        learning_rate=learning_rate,
        reg_lambda=reg_lambda,
        reg_alpha=reg_alpha,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        max_depth=max_depth,
    )
    model.fit(xtrain, ytrain, early_stopping_rounds=300, eval_set=[(xvalid, yvalid)], verbose=1000)
    preds_valid = model.predict(xvalid)
    rmse = mean_squared_error(yvalid, preds_valid, squared=False)
    return rmse
```

```python
def transformdf(df):
    
    df.drop(columns=['Piso', 'Posición', 'Número de frentes', 'Valor comercial (USD)'], axis=1, inplace=True)
    df['year'] = df['Fecha entrega del Informe'].dt.year
    df['month'] = df['Fecha entrega del Informe'].dt.month
    df.drop(columns='Fecha entrega del Informe', axis=1, inplace=True)

    
    df['Área Terreno'] = df['Área Terreno'].apply(lambda x: str(x).replace(',', '')) 
    df['Área Construcción'] = df['Área Construcción'].apply(lambda x: str(x).replace(',', '')) 
    df['Área Terreno'] = pd.to_numeric(df['Área Terreno'], errors='coerce')
    df['Área Construcción'] = pd.to_numeric(df['Área Construcción'], errors='coerce')
    
    df['Departamento'] = df['Departamento'].apply(lambda x: str(x).lower()) 
    df['Provincia'] = df['Provincia'].apply(lambda x: str(x).lower()) 
    df['Distrito'] = df['Distrito'].apply(lambda x: str(x).lower())
    
    df['Departamento'] = df['Departamento'].apply(strip_accents) 
    df['Provincia'] = df['Provincia'].apply(strip_accents) 
    df['Distrito'] = df['Distrito'].apply(strip_accents)
    
    df["Tipo de vía"].fillna(0, inplace = True)
    df["Número de estacionamiento"].fillna(0, inplace = True)
    df["Número de estacionamiento"].fillna(0, inplace = True)
    dist_coord_lost = df[((df['Latitud (Decimal)'].isnull() == True) | (df['Longitud (Decimal)'].isnull() == True)) == True]
    mask_idx = list(dist_coord_lost.index)
    for i in mask_idx:
        distrito = df.loc[i, 'Distrito']
        provincia = df.loc[i, 'Provincia']

        if distrito in dist_sh:
            df.loc[i, 'Latitud (Decimal)'] = (distritos_cent.loc[distritos_cent['NOMBDIST'] == distrito, 'lat_cent']).values[0]
            df.loc[i, 'Longitud (Decimal)'] = (distritos_cent.loc[distritos_cent['NOMBDIST'] == distrito, 'lon_cent']).values[0]
        elif provincia in prov_sh:
            df.loc[i, 'Latitud (Decimal)'] = (provincia_cent.loc[provincia_cent['NOMBPROV'] == provincia, 'lat_cent']).values[0]
            df.loc[i, 'Longitud (Decimal)'] = (provincia_cent.loc[provincia_cent['NOMBPROV'] == provincia, 'lon_cent']).values[0]
        else:
            df.loc[i, 'Latitud (Decimal)'] = -9.19
            df.loc[i, 'Longitud (Decimal)'] = -75.016
    
    
    Edad = []
    for i in df["Edad"]:
      if i > 1600:
        Edad.append(2022-i)
      else:
        Edad.append(i)
    df['Edad'] = Edad
    df["Edad"].fillna(np.mean(df["Edad"]), inplace = True)
    df['Categoría del bien'].fillna('Otro', inplace = True)
    dtype0 = {'Tipo de vía': 'category',
                            'Departamento': 'category',
                            'Provincia': 'category',
                            'Distrito': 'category',
                            'Categoría del bien': 'category',
                            'Estado de conservación': 'category',
                            'Método Representado': 'category'}
    df =df.astype(dtype0)
    object_cols = ['Tipo de vía', 'Departamento', 'Provincia', 'Distrito', 'Categoría del bien', 'Estado de conservación', 'Método Representado']
    
    xtest = df[useful_features]
    xtest[object_cols] = ordinal_encoder.transform(xtest[object_cols])
    
    return xtest
```
